**Importing Libraries**

```python
import pandas as pd
from mlxtend.preprocessing import TransactionEncoder
from mlxtend.frequent_patterns import apriori, fpgrowth, association_rules
```

**Loading the dataset**

```python
dataset = pd.read_csv("TRAIN-ARULES.csv")
dataset.product_name = ["".join(i.split(",")) for i in dataset.product_name]
order_labels = dataset.order_id.unique()
```

**Preprocessing the dataset**

```python
df1 = [dataset[dataset.order_id == i]["product_name"].tolist() for i in order_labels]
trans_encoder = TransactionEncoder().fit(df1)
df1 = trans_encoder.transform(df1)
df1 = pd.DataFrame(df1,columns=trans_encoder.columns_)
```

**Apply fpgrowth and and apriori algorithm and finding association_rules**

```python
apriori_items = apriori(df1, min_support = 0.0045, use_colnames = True)
apriori_items_fpgrowth = fpgrowth(df1, min_support = 0.0045, use_colnames=True)
association_rules_apriori = association_rules(apriori_items, metric = "confidence")
association_rules_fpgrowth = association_rules(apriori_items_fpgrowth, metric = "confidence")
```

```python
print("Association Rules build using Apriori Method:- \n")
for item in association_rules_apriori.iterrows():
  print(f"Rule {item[0] + 1}: {tuple(item[1].antecedents)} -> {tuple(item[1].consequents)}")
  # print(item[1])
```

```python
print("Association Rules build using fpgrowth Method:- \n")
for item in association_rules_fpgrowth.iterrows():
  print(f"Rule {item[0] + 1}: {tuple(item[1].antecedents)} -> {tuple(item[1].consequents)}")
  # print(item[1])
```
